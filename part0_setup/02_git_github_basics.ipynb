# Git & GitHub Basics for ML Projects

## Objective
Understand and document core Git workflows used in machine learning projects,
including versioning experiments, notebooks, and collaboration-ready code.

## Why Git matters in ML
- Track experiments and changes
- Revert broken models or pipelines
- Collaborate with other engineers
- Maintain a clean project history

```python
import subprocess

git_version = subprocess.run(
    ["git", "--version"],
    capture_output=True,
    text=True
)

print(git_version.stdout)
```

```python
status = subprocess.run(
    ["git", "status"],
    capture_output=True,
    text=True
)

print(status.stdout)
```
