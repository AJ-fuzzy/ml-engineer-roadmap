## Automatic Differentiation (Autograd)

PyTorch automatically tracks operations on tensors
to compute gradients for optimization.

```python
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1

y.backward()

print("y:", y.item())
print("dy/dx:", x.grad)
```

```python
with torch.no_grad():
    z = x ** 2

z
```
