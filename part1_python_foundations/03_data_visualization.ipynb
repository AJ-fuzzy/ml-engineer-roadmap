# Data Visualization with Matplotlib and Seaborn

## Objective
Learn and document basic data visualization techniques for ML:
- Line plots, bar plots, scatter plots
- Histograms and distributions
- Customizing plots
- Using Seaborn for high-level visualizations

## Why this matters
Visualization is critical for:
- Exploring datasets
- Detecting patterns and outliers
- Communicating results effectively


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Optional: nicer plots
sns.set_style("whitegrid")
%matplotlib inline
```

```python
# Sample dataset
data = {
    "Day": np.arange(1, 8),
    "Temperature": [22, 21, 23, 24, 25, 23, 22],
    "Humidity": [30, 35, 32, 31, 29, 33, 34]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,4))
plt.plot(df["Day"], df["Temperature"], marker='o', label="Temperature (C)")
plt.plot(df["Day"], df["Humidity"], marker='s', label="Humidity (%)")
plt.title("Temperature and Humidity Over a Week")
plt.xlabel("Day")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(df["Day"], df["Temperature"], color='skyblue')
plt.title("Temperature per Day")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# Generate random data
np.random.seed(0)
scores = np.random.randint(50, 100, 50)

plt.hist(scores, bins=10, color='salmon', edgecolor='black')
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot using Seaborn
sns.scatterplot(x="Temperature", y="Humidity", data=df, s=100, color='green')
plt.title("Temperature vs Humidity")
plt.show()
```

```python
# Create a bigger dataset for demonstration
df_large = pd.DataFrame({
    "Height": np.random.normal(170, 10, 50),
    "Weight": np.random.normal(65, 8, 50),
    "Age": np.random.randint(18, 50, 50)
})

sns.pairplot(df_large)
plt.show()
```
